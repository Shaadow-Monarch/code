```python
import pandas as pd
```

## **BATTING RECORDS OF INDIAN CRICKET TEAM** in ODI

```python
url = "https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?class=2;id=6;type=team"
```

```python
df = pd.read_html(url,header = 0)
df
```

###How many tables are there in webpages

```python
len(df)
```

###tables content after filtering

```python
df_new = df[0]
df_new = df_new[["Player","100"]]
df_new 
```

###shape and size of table

```python
df_new.shape
```

###changing in the name of column

```python
df_new.rename(columns = {"100":"century"}, inplace = True)
df_new
```

##Quick Exploratory Data Analysis

###On basis of Century

```python
import seaborn as sns
```

```python
p = sns.distplot(df_new.century,kde=False,hist_kws=dict(edgecolor="black", linewidth=2))
p.set(ylabel = " NO. of Players")
```

###on basis of Runs


```python
df_new2 = df[0]
p = sns.distplot(df_new2.Runs,kde=False,hist_kws=dict(edgecolor="black", linewidth=2))
p.set(ylabel = "NO. of Players")
```
